**Importacion de librerias**

```python
import pandas as pd
import sqlalchemy
import psycopg2
from time import time
```

**Lectura del archivo .csv mediante la libreria panda definida como pd**

```python
%%time
df = pd.read_csv("Supermarket sales.csv",delimiter=';')
```

**Cambio de encabezados mediante un diccionario con pandas**

```python
%%time
df = df.rename(columns={'Invoice ID': 'ID factura','Branch': 'Rama','City': 'Ciudad','Customer type': 'Tipo cliente',
                        'Gender': 'Genero','Product line': 'Linea de productos','Unit price': 'Precio unitario',
                        'Quantity': 'Cantidad','Tax 5%': 'Impuesto 5%','Date': 'Fecha','Time': 'Hora','Payment': 'Pago',
                        'cogs': 'Costo de vienes vendidos','gross margin percentage': 'Porcentaje de margen bruto',
                        'gross income': 'Ingresos brutos','Rating': 'Calificación'})
df
```

**Traducción de datos al español**

```python
%%time
df["Pago"].replace({"Ewallet":"Monedero digital","Credit card":"Tarjeta de credito","Cash":"Efectivo" },inplace=True)
df["Genero"].replace({"Female":"Femenino","Male":"Mascuino"},inplace=True)
df["Linea de productos"].replace({"Health and beauty":"Salud y belleza","Electronic accessories":"Accesorios electrónicos",
                                 "Home and lifestyle":"Hogar y estilo de vida","Sports and travel":"Deportes y viajes",
                                 "Fashion accessories":"Accesorios de moda","Food and beverages":"Comida y bebidas"},inplace=True)
df["Tipo cliente"].replace({"Member":"Miembro"},inplace=True)
df
```

# Limpieza de datos y cambio de datos #

**Calculo de impuesto**

```python
%%time
df['Impuesto 5%']=df['Precio unitario']* df['Cantidad']*0.05
df
```

**Calculo del total**

```python
%%time
df['Total']=df['Precio unitario']* df['Cantidad'] + df['Impuesto 5%']
df
```

**Calculo de costos de vienes vendidos**

```python
%%time
df['Costo de vienes vendidos']=df['Precio unitario']* df['Cantidad']
df
```

**verificación porcentaje de margen bruto**

```python
df['Porcentaje de margen bruto'].unique()
```

**cambio de valor porcentaje de margen bruto**

```python
df['Porcentaje de margen bruto'] = pd.unique(df['Porcentaje de margen bruto']) [0]
df
```

**Cambio de ingresos brutos**

```python
%%time
df['Ingresos brutos']=df['Costo de vienes vendidos']* (df['Porcentaje de margen bruto']/100)
df
```

**verificacion de calificaciones**

```python
df['Calificación'].unique().tolist()
```

**Cargando datos en PostgreSQL**

```python
%%time
engine = sqlalchemy.create_engine('postgresql://postgres:postgres@localhost:5432/Taller1')

df.to_sql('ventas',engine,if_exists='replace')
```
